# AMAZON ALEXA REVIEW RATINGS CLASSIFICATION

#STEP #1: PROBLEM STATEMENT

Dataset consists of 3000 Amazon customer reviews, star ratings, date of review, variant and feedback of various amazon Alexa products like Alexa Echo, Echo dots.
The objective is to discover insights into consumer reviews and perfrom sentiment analysis on the data.
Dataset: www.kaggle.com/sid321axn/amazon-alexa-reviews

# Step 2 : Importing data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('amazon_alexa.tsv',sep='\t')
data.head()
```

```python
data.verified_reviews
```

```python
data=data.drop(['date','rating'],axis=1)
```

```python
data
```

```python
dummies=pd.get_dummies(data.variation,drop_first=True)
```

```python
dummies
```

```python
data.drop(['variation'],axis=1,inplace=True)
```

```python
data=pd.concat([data,dummies],axis=1)
```

```python
data
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
```

```python
data.verified_reviews=enc.fit_transform(data.verified_reviews)
```

```python
data.info()
```

```python
X=data.drop(['feedback'],axis=1)
y=data.feedback
```

```python
y
```

```python
X
```

# Model training

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=.2,random_state=10)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=100,criterion='entropy')
model.fit(X_train,y_train)
```

# Evaluating the model

```python
y_predict= model.predict(X_test)
```

```python
accuracy_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```

```python
cm=confusion_matrix(y_test,y_predict)
sns.heatmap(cm,annot=True,fmt='d')
```

```python
pd.crosstab(y_test,y_predict)
```
